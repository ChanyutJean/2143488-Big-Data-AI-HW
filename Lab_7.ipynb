```python
# Don't forget to import data!
import pandas as pd
df = pd.read_csv("W8_HW_dataset.csv", header=None, names=list("ABCDEFGHIJKLMNOP"))
```

```python
print(df.head())
print(df.tail())
print(df.info())
# print(df.describe())
# print(df.isna())
```

```python
# It took 1-2 minutes for me.
T = []
for id in range(len(df)):
  T.append([item for item in df.loc[id] if not pd.isnull(item)])
T
```

```python
from mlxtend.preprocessing import TransactionEncoder

encoder=TransactionEncoder()
encoder_arr=encoder.fit_transform(T)
df2=pd.DataFrame(encoder_arr,columns=encoder.columns_)
df2
```

```python
df2.sum(axis=0)
```

```python
from mlxtend.frequent_patterns import fpgrowth

frequent_itemsets1 = fpgrowth(df2, min_support=0.10, use_colnames=True)
frequent_itemsets1['length'] = frequent_itemsets1['itemsets'].apply(lambda x: len(x))
frequent_itemsets1
```

```python
frequent_itemsets1[frequent_itemsets1['length'] > 1].sort_values("support", ascending=False)
```
